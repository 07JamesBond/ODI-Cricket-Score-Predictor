```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
import ipywidgets as widgets
from IPython.display import display

# Load your dataset
data = pd.read_csv('/content/drive/MyDrive/CSE-303 Statistical Analysis for Data Science/Project/Dataset/odi.csv')

# Encode categorical features
label_encoder = LabelEncoder()
data["venue"] = label_encoder.fit_transform(data["venue"])
data["bat_team"] = label_encoder.fit_transform(data["bat_team"])
data["bowl_team"] = label_encoder.fit_transform(data["bowl_team"])

# Select relevant features for prediction
X = data[["venue", "bat_team", "bowl_team", "overs", "runs_last_5", "wickets_last_5"]]

# Target variable
y_runs = data["total"]

# Train the Random Forest regressor for score prediction
rf_regressor = RandomForestRegressor(n_estimators=200, random_state=42)
rf_regressor.fit(X, y_runs)

def predict_score(venue, bat_team, bowl_team, overs, runs, wickets, model):
    encoded_venue, encoded_bat_team, encoded_bowl_team = update_encoded_values(venue, bat_team, bowl_team)

    input_data = pd.DataFrame(columns=X.columns)
    input_data.loc[0] = 0
    input_data.loc[0, 'venue'] = encoded_venue
    input_data.loc[0, 'bat_team'] = encoded_bat_team
    input_data.loc[0, 'bowl_team'] = encoded_bowl_team
    input_data.loc[0, 'overs'] = overs
    input_data.loc[0, 'runs_last_5'] = runs
    input_data.loc[0, 'wickets_last_5'] = wickets

    predicted_score = model.predict(input_data)[0]
    print(f"Predicted Score: {int(predicted_score)}")

def update_encoded_values(venue, bat_team, bowl_team):
    encoded_venue = label_encoder.transform([venue])[0]
    encoded_bat_team = label_encoder.transform([bat_team])[0]
    encoded_bowl_team = label_encoder.transform([bowl_team])[0]
    return encoded_venue, encoded_bat_team, encoded_bowl_team

batting_team_options = data["bat_team"].unique()
bowling_team_options = data["bowl_team"].unique()
venue_options = data["venue"].unique()

venue = widgets.Dropdown(options=venue_options, description='Venue:')
bat_team = widgets.Dropdown(options=batting_team_options, description='Batting Team:')
bowl_team = widgets.Dropdown(options=bowling_team_options, description='Bowling Team:')
overs = widgets.FloatSlider(min=0, max=50, value=10, description='Overs:')
runs = widgets.FloatText(placeholder='Runs', description='Runs:')
wickets = widgets.IntSlider(min=0, max=9, value=0, description='Wickets:')

predict_button = widgets.Button(description='Predict Score')

# Define a function to handle button click
def on_predict_button_click(b):
    selected_bat_team = bat_team.value
    selected_bowl_team = bowl_team.value
    selected_venue = venue.value
    selected_overs = overs.value
    selected_runs = runs.value
    selected_wickets = wickets.value

    predict_score(selected_venue, selected_bat_team, selected_bowl_team, selected_overs, selected_runs, selected_wickets, rf_regressor)

predict_button.on_click(on_predict_button_click)

# Display widgets
display(venue, bat_team, bowl_team, overs, runs, wickets, predict_button)
```
